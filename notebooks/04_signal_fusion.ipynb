This notebook demonstrates the final step of the ORACLE-26 pipeline: fusing signals from all four verticals into a single, high-density predictive ranking.

```python
import os
import pandas as pd
import numpy as np
import sys
from pathlib import Path

sys.path.append(str(Path(os.getcwd()).parent))

from src.features.fusion import SignalFusionEngine, SignalVector

fusion = SignalFusionEngine()
print("Fusion engine initialized.")
```

### 1. The Signal Vector
We represent each team as a 5-dimensional vector: [Sports, Markets, Finance, Climate, Social].

```python
example_sv = SignalVector(
    subject="Argentina",
    vertical="wc2026",
    sports=0.95,
    markets=0.88,
    finance=0.60,
    climate=0.82,
    social=0.90
)

result = fusion.fuse(example_sv)
print(f"Fused Conflux Score for Argentina: {result.conflux_score:.4f}")
print(f"Interpretation: {result.interpretation}")
```
