### 2. Alpha Calculation
Detecting where the model view diverges from the market view.

```python
conflux_df = pd.read_csv("../data/processed/conflux_wc2026.csv")
conflux_df['alpha_gap'] = conflux_df['sports'] - conflux_df['markets']

print("Top 5 VALUE opportunities (Model > Market):")
conflux_df.sort_values("alpha_gap", ascending=False).head(5)[['subject', 'sports', 'markets', 'alpha_gap']]
```

```python
print("Top 5 HYPE warnings (Market > Model):")
conflux_df.sort_values("alpha_gap", ascending=True).head(5)[['subject', 'sports', 'markets', 'alpha_gap']]
```
